# kMeans demo

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs, make_circles, make_moons, make_swiss_roll

from sklearn.cluster import KMeans
```

```python
# Clusters exampels
def plot_cluster (attributes, clusters):
    plt.scatter(attributes[:, 0], attributes[:, 1], c = clusters)
```

```python
FILE_NAME = 'Datasets/unbalance/unbalance.txt'
points = pd.read_csv(FILE_NAME, sep=' ', header=None, names = ['x', 'y'])
```

```python
points
```

```python
points.shape
```

```python
points.describe()
```

```python
plt.scatter(points['x'], points['y'])
```

```python
# kMeans clustering with random restart
# average complexity: O(k n T), where n is the number of samples and T is the number of iteration
K = 8
k_means = KMeans(n_clusters = K, init = "random", n_init = 10, max_iter = 300)
k_means.fit(points)
assigned = k_means.predict(points)
plt.scatter(points['x'],  points['y'], c = assigned)
```

```python
# kMeans++
k_means_plus_plus = KMeans(n_clusters = K, init = "k-means++", max_iter = 300, n_init = 'auto')
clusters = k_means_plus_plus.fit_predict(points)
plt.scatter(points['x'], points['y'], c = clusters)
plt.show()

k_means_plus_plus.cluster_centers_
```

```python
# kMeans score test
test_data, test_clusters = make_blobs(n_samples = 200)

k_means_for_score = KMeans(3, n_init = 'auto')
assigned_clusters = k_means_for_score.fit_predict(test_data)

plt.title("Original clusters")
plot_cluster(test_data, test_clusters)
plt.show()

plt.title('Predicted clusters')
plot_cluster(test_data, assigned_clusters)
plt.show()
```

```python
# kMeans problems
circle_attributes, circle_clusters = make_circles()
plot_cluster(circle_attributes, circle_clusters)
plt.title('Original clusters')
plt.show()

circle_assigned_clusters = KMeans(2, n_init = 'auto').fit_predict(circle_attributes)
plot_cluster(circle_attributes, circle_assigned_clusters)
plt.title('Predicted clusters')
plt.show()
```

```python
# kMeans problems
moons_attributes, moons_clusters = make_moons(noise = 0.05)
plot_cluster(moons_attributes, moons_clusters)
plt.title('Original clusters')
plt.show()

moons_assigned_clusters = KMeans(2, n_init = 'auto').fit_predict(moons_attributes)
plot_cluster(moons_attributes, moons_assigned_clusters)
plt.title('Predicted clusters')
plt.show()
```

```python
def plot3d(sr_points, sr_color):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection="3d")
    fig.add_axes(ax)
    ax.scatter(
        sr_points[:, 0], sr_points[:, 1], sr_points[:, 2], c=sr_color, s=50, alpha=0.8
    )
    ax.set_title("Swiss Roll")
    ax.view_init(azim=-66, elev=12)
    _ = ax.text2D(0.8, 0.05, s="1500 samples", transform=ax.transAxes)
```

```python
sr_attributes, sr_clusters = make_swiss_roll(n_samples=1500, random_state=0)
plot3d(sr_attributes, sr_clusters)
```

```python
sr_assigned_clusters = KMeans(3, n_init = 'auto').fit_predict(sr_attributes)
plot3d(sr_attributes, sr_assigned_clusters)
```
