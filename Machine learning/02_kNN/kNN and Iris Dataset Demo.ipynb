## *k*-Nearest Neighbors (*k*NN)
- Supervised learning
    - Classification
    - Regression
- Instance-based learning (Lazy learning)
- The simplest algorithm possible.

Saves the entire dataset (usually it is using k-d tree). To classify a new element, it finds the closest _k_ elements (linearly, in Euclidean space) to it and responds with the most common class:

```python
import matplotlib.pyplot as plt
import mglearn
from IPython.display import display

%matplotlib notebook
```

```python
from sklearn.datasets import make_blobs
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

The number of neighbors determines how "smooth" the boundary between the two classes is. Here is an example with a different number of neighbors:

```python
from sklearn.neighbors import KNeighborsClassifier

X, y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} neighbour(s)".format(n_neighbors))
```

The background in the above diagram determines which of the two classes a point will fall into. Note that for k=1 the boundary is broken and captures every element. For larger k, the boundary becomes smoother, although it misclassifies some elements. This can be an advantage (it will ignore anomalies in the data).

## *k*NN & Iris Dataset
Below is an example with the __Iris__ dataset and an implementation of the __*k*NN__ algorithm taken from the __"sklearn"__ library.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn import datasets
from sklearn.model_selection import train_test_split , KFold
from sklearn.preprocessing import Normalizer
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

from collections import Counter
```

#### Creating a DataFrame object.
The data frames are objects used in Python and R to represent a table or a two-dimensional array-like structure in which each column contains values of one variable and each row contains one set of values from each column. They serve to present the dataset in an easy and understandable way.

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                      columns= iris['feature_names'] + ['target'])
iris_df.head()
```

```python
iris_df
```

```python
iris_df.describe()
```

Usually most libraries as well as _"sklearn"_ want the dataset to be split into features and class. We mark them with ___x___ and ___y___, respectively.

```python
x = iris_df.iloc[:, :-1]
y = iris_df.iloc[:, -1]
```

```python
x.head()
```

```python
y.head()
```

Also the dataset is splitted into training and test datasets.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True, random_state=0)

x_train = np.asarray(x_train)
y_train = np.asarray(y_train)

x_test = np.asarray(x_test)
y_test = np.asarray(y_test)
```

```python
print(f'training set size: {x_train.shape[0]} samples \ntest set size: {x_test.shape[0]} samples')
```

It is always better to normalize the data!

```python
scaler = Normalizer().fit(x_train)
normalized_x_train = scaler.transform(x_train)
normalized_x_test = scaler.transform(x_test)
```

```python
print('x train before Normalization')
print(x_train[0:5])
print('\nx train after Normalization')
print(normalized_x_train[0:5])
```

Creating a DataFrame object with the normalized data and making some visualizations.

```python
di = {0.0: 'Setosa', 1.0: 'Versicolor', 2.0:'Virginica'}
iris_df_2= pd.DataFrame(data= np.c_[normalized_x_train, y_train],
                        columns= iris['feature_names'] + ['target'])
```

```python
before = sns.pairplot(iris_df.replace({'target': di}), hue= 'target', hue_order = ['Setosa', 'Versicolor', 'Virginica'])
before.fig.suptitle('Pair Plot of the dataset Before normalization', y=1.08)
```

```python
after = sns.pairplot(iris_df_2.replace({'target':di}), hue= 'target', hue_order = ['Setosa', 'Versicolor', 'Virginica'])
after.fig.suptitle('Pair Plot of the dataset After normalization', y=1.08)
```

#### An example with *k*NN using *k*=3

```python
K=3
```

```python
knn = KNeighborsClassifier(K)
knn.fit(normalized_x_train, y_train)
y_pred_sklearn = knn.predict(normalized_x_test)
print(y_pred_sklearn)
```

```python
print(f'The accuracy of sklearn implementation is {accuracy_score(y_test, y_pred_sklearn)}')
```

Basically, the datasets are training, validation and test, but when the data is not much, cross-validation is used. An example with 10-fold cross-validation can be seen below.

```python
n_splits = 10
kf= KFold(n_splits = n_splits)

accuracy_k = []
k_values= list(range(1,30,1))

for k in k_values:
    accuracy_fold = 0
    for normalized_x_train_fold_idx, normalized_x_valid_fold_idx in  kf.split(normalized_x_train):
        normalized_x_train_fold = normalized_x_train[normalized_x_train_fold_idx]
        y_train_fold = y_train[normalized_x_train_fold_idx]
        
        normalized_x_test_fold = normalized_x_train[normalized_x_valid_fold_idx]
        y_valid_fold = y_train[normalized_x_valid_fold_idx]
        
        knn=KNeighborsClassifier(k)
        knn.fit(normalized_x_train_fold, y_train_fold)
        y_pred_fold = knn.predict(normalized_x_test_fold)
        
        accuracy_fold += accuracy_score (y_pred_fold, y_valid_fold)
    accuracy_fold = accuracy_fold/ n_splits
    accuracy_k.append(accuracy_fold)
```

```python
print(f'The accuracy for each K value was {list ( zip (accuracy_k, k_values))}')
```

```python
print(f'Best accuracy was {np.max(accuracy_k)}, which corresponds to a value of K= {k_values[np.argmax(accuracy_k)]}')
```

The function ___np.argmax___ will give the first ___k___ with maximum accuracy. In order to make an accurate analysis and choose the best ___k___, we will draw a graph showing how the accuracy varies with the value of the parameter ___k___.

```python
k2acc = {'k': k_values, 'acc': accuracy_k}
k2acc = pd.DataFrame(k2acc)
```

```python
sns.set(style='darkgrid')
ax = sns.lineplot(x='k', y='acc', data=k2acc, marker='o')
ax.set(xticks=k_values)
plt.show()
```
